```
import random
import numpy as np
import matplotlib.pyplot as plt
j_list=[]
head_l=[]
tail_l=[]
ph_l=[]
pt_l=[]
for i in range(1,51):
  j=10*i
  j_list.append(j)
  head=0
  tail=0
  for k in range(0,j):
    value=random.randint(1,100000)
    if value%2==0:
      head+=1
    else:
      tail+=1
  ph =head/j
  pt=tail/j
  j_arr=np.array(j_list)# converting list into array
  head_l.append(head)# appending values into list
  tail_l.append(tail)
  ph_l.append(ph)
  pt_l.append(pt)
  print(head,tail,round(ph,2),round(pt,2),sep=' ')  #rounding off probabilities to two decimal places
  
```

```
ph_arr=np.array(ph_l)
pt_arr=np.array(pt_l)#converting list into array
plt.plot(j_arr,ph_arr,label="P(H)")
plt.plot(j_arr,pt_arr,label="P(T)")
plt.xlabel("NO.OF TOSSES")
plt.ylabel("PROBABILITIES")
plt.title("Probability vs No.of tosses")
plt.legend()
plt.show()
```
